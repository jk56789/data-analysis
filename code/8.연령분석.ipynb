```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as img
import pandas as pd
df = pd.read_csv('C:/Users/dlwnr/a/thesis_code/use.csv', index_col=0)
df
```

```python
age_group = []
for row in df['age']:
    if row >= 70:
        age_group.append('70대 이상')
    elif row >= 60:
        age_group.append('60대')
    elif row >= 50:
        age_group.append('50대')
    elif row >= 40:
        age_group.append('40대')
    elif row >= 30:
        age_group.append('30대')
    elif row >= 20:
        age_group.append('20대')
    elif row >= 10:
        age_group.append('10대')

df['age_group'] = age_group
df
```

welch-anova (이분산성 이므로)

```python
import pingouin as pg

pg.welch_anova(dv = 'sad', between = 'age_group', data = df)

#p값이 작다. 평균의 적어도 하나 다르다.
```

사후분석

```python
post=pg.pairwise_gameshowell(dv = 'sad', between = 'age_group', data = df)
post
```

```python
from statsmodels.sandbox.stats.multicomp import MultiComparison
import scipy.stats

comp = MultiComparison(df.sad, df.age_group)
result = comp.allpairtest(scipy.stats.ttest_ind, method='bonf')
result[0]
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

hsd = pairwise_tukeyhsd(df['sad'], df['age_group'], alpha=0.05)
hsd.summary()
```

```python
fig = hsd.plot_simultaneous()
plt.show()
```

```python
import statsmodels.formula.api as sm 


model = sm.ols(formula = 'sad ~ age',data=df).fit() 
model.summary()
```
