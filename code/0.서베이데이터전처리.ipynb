```python
import pandas as pd
df = pd.read_csv('C:/Users/dlwnr/graduation_thesis/data/use_data/2021년-서울서베이-시민_데이터.csv')
df
```

```python
#sns

# 열을 추가하자(특졍열에 맞춰서)
# 아래는 df['Q11]이 어떤 값 일때 df['Q15']를 원하는 값으로 변경해주었다.

df.loc[df['Q11'] == 8,'Q15'] = 0
df.loc[df['Q11'] == 1,'Q15']= 1.000000
df.loc[df['Q11'] == 2,'Q15'] = 0.142857
df.loc[df['Q11'] == 3,'Q15'] = 0.033333
df.loc[df['Q11'] == 4,'Q15'] = 0.005555
df.loc[df['Q11'] == 5,'Q15'] = 0.002740

# #NULL!값을 0으로 바꿔주자
my_list=df['Q11_1']

for index, value in enumerate(my_list):
    if value == '#NULL!' :
      my_list[index] = 0


# #NULL!바꾸고 int형으로 변환
df['Q11_1'] = list(map(int, my_list))

df['raw_count']=df['Q11_1']*df['Q15']
```

```python
#연령
df['new_age'] = 2021
df['new_age'] = df['new_age'] - df['YEAR']
```

```python
#연령 그룹
age_group = []
for row in df['new_age']:
    if row >= 70:
        age_group.append('70대 이상')
    elif row >= 60:
        age_group.append('60대')
    elif row >= 50:
        age_group.append('50대')
    elif row >= 40:
        age_group.append('40대')
    elif row >= 30:
        age_group.append('30대')
    elif row >= 20:
        age_group.append('20대')
    elif row >= 10:
        age_group.append('10대')

df['age_group'] = age_group
```

```python
#돈


#상승
df.loc[((df['Q24_1']==1) & (df['Q24_2A'] == '1.00'))|((df['Q24_1']==2) & (df['Q24_2B'] == '1.00')), 'Q15'] = '상승'
#하강
df.loc[(df['Q24_1']==1) & (df['Q24_2A'] == '2.00')|((df['Q24_1']==2) & (df['Q24_2B'] == '2.00')),'Q15']= '하강'
#아웃
df.loc[(df['Q24_1']==1) & (df['Q24_2A'] == '3.00')|((df['Q24_1']==2) & (df['Q24_2B'] == '3.00')),'Q15'] = '해고 및 폐업'
#변화없음
df.loc[(df['Q24_1']==1) & (df['Q24_2A'] == '4.00')|((df['Q24_1']==2) & (df['Q24_2B'] == '4.00'))|(df['Q24_1']==3),'Q15'] = '변호없음'
```

```python
#학력

#고등학교
df.loc[(df['DQ1_1']==1) | (df['DQ1_1']==2) |(df['DQ1_1']==3) | (df['DQ1_1']==4), 'DQ1_2'] = '고등학교'
#대학교
df.loc[(df['DQ1_1']==5)|(df['DQ1_1']==6),'DQ1_2']= '대학교'
#대학원
df.loc[(df['DQ1_1']==7),'DQ1_2'] = '대학원'
```

```python
# 성별

#남
df.loc[(df['SEX']==1) , 'Q3_1_1'] = '남'
#여
df.loc[(df['SEX']==2) , 'Q3_1_1'] = '여'
```

```python
# 재택근무


df['Q20'].value_counts()

# #NULL!값을 0으로 바꿔주자
jh_list=df['Q20']

for index, value in enumerate(jh_list):
    if value == '#NULL!' :
      jh_list[index] = 0


# #NULL!바꾸고 int형으로 변환
df['Q20'] = list(map(int, jh_list))
df['Q20'].value_counts()

#재택근무 경험 있다.
df.loc[(df['Q20']==1) , 'Q3_1_2'] = '경험 있음'
#재택근무 경험 없다.
df.loc[(df['Q20']==2) , 'Q3_1_2'] = '경험 없음'
#일을 하지 않는다.
df.loc[(df['Q20']==0) , 'Q3_1_2'] = '일을 안함'


```
