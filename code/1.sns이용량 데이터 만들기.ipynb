```python
#데이터 프레임 준비
import pandas as pd
df = pd.read_csv('C:/Users/dlwnr/graduation_thesis/data/use_data/2021년-서울서베이-시민_데이터.csv')
df=df[['Q22','Q11','Q11_1','Q15']] #우울지수,sns 사용량 기준,사용량 기준 사용 횟수,아무거나 열
df
```

```python
# 열을 추가하자(특졍열에 맞춰서)
# 아래는 df['Q11]이 어떤 값 일때 df['Q15']를 원하는 값으로 변경해주었다.

df.loc[df['Q11'] == 8,'Q15'] = 0
df.loc[df['Q11'] == 1,'Q15']= 1.000000
df.loc[df['Q11'] == 2,'Q15'] = 0.142857
df.loc[df['Q11'] == 3,'Q15'] = 0.033333
df.loc[df['Q11'] == 4,'Q15'] = 0.005555
df.loc[df['Q11'] == 5,'Q15'] = 0.002740
df

```

```python
# 데이터 프레임의 원소를  int형으로 바꾸려 했으나 오류가 아래와 같이 나왔다.
# #NULL!값을 0으로 바꿔주자(사용한 데이터에서는 실제로 0을 의미)

df['Q11_1'] = list(map(int, df['Q11_1']))
df
```

```python
# #NULL!값을 0으로 바꿔주자
my_list=df['Q11_1']

for index, value in enumerate(my_list):
    if value == '#NULL!' :
      my_list[index] = 0


my_list.value_counts()
```

```python
# #NULL!바꾸고 int형으로 변환
df['Q11_1'] = list(map(int, my_list))
df
```

```python
# 1일 sns 사용 횟수를 알게 되었다.
df['raw_count']=df['Q11_1']*df['Q15']
df['raw_count'].describe()
```

```python
sns_group = []
for row in df['raw_count']:
    if row >= 10:
        sns_group.append('10회 이상')
    elif row >= 5:
        sns_group.append('5회 이상')
    elif row >= 3:
       sns_group.append('3회 이상')
    elif row >  0:
        sns_group.append('1회 이상')
    elif row ==  0:
        sns_group.append('사용하지 않음')
    

df['sns_group'] = sns_group
df['sns_group'].value_counts()
```
